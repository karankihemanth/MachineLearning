```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("fraudTrain.csv")
```

```python
df.head()
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
df=df.drop(columns=["Unnamed: 0"],axis=1)
```

```python
df.head()
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])
df['amt'] = df['amt'].astype(float)
df['is_fraud'] = df['is_fraud'].astype(int)
```

```python
df['gender'] = df['gender'].astype('category')
df['category'] = df['category'].astype('category')
df['job'] = df['job'].astype('category')
```

```python
df['trans_date_trans_time'] = df['trans_date_trans_time'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

columns_to_encode = [
    'trans_date_trans_time', 'merchant', 'category', 'first', 'last',
    'gender', 'street', 'city', 'state', 'job', 'dob', 'trans_num'
]
for column in columns_to_encode:
    encoder.fit(df[column])
    df[column] = encoder.transform(df[column])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

categorical_cols = ['merchant', 'category', 'first', 'last', 'job', 'gender', 'street', 'city']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python

X = df.drop(columns=['is_fraud', 'trans_date_trans_time', 'dob', 'cc_num', 'trans_num'])
y = df['is_fraud']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy_score   :",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred, labels=model.classes_)
cm
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='afmhot_r')
```
